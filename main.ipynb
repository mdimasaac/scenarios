```python
import pandas as pd
df1 = pd.read_excel("HR Data.xlsx")
df2 = pd.read_csv("food_wastage_data.csv")
df3 = pd.read_csv("LA crime - processed.csv")
```

```python
print(df1.shape)
display(df1.head())

print(df2.shape)
display(df2.head())

print(df3.shape)
display(df3.head())
```

```python
df3 = df3.sample(n = 1500)
print(df3.shape)
display(df3.head())
```

```python
df3.dtypes
```

```python
df3["date_reported"] = pd.to_datetime(df3["date_reported"])
```

```python
df3 = df3.sort_values(by ="date_reported")
display(df3.head())
```

```python
display(df1.isna().sum())
display(df2.isna().sum())
display(df3.isna().sum())
```

```python
df3.to_csv("crime_data.csv", index = False)
df1.to_csv("employee_data.csv", index = False)
```
